# Download APK Dataset

```python
import os

url = 'https://androzoo.uni.lu/static/lists/latest_with-added-date.csv.gz'
filename = os.path.join(os.path.abspath(''), 'APKs', 'latest_with-added-date.csv.gz')
```

### Download

```python
from urllib.request import urlretrieve
import progressbar

pbar = None

def show_progress_bar(block_num, block_size, total_size):
    global pbar
    if pbar is None:
        pbar = progressbar.ProgressBar(max_value=total_size)
        pbar.start()

    downloaded = block_num * block_size
    if downloaded < total_size:
        pbar.update(downloaded)
    else:
        pbar.finish()
        pbar = None

if os.path.exists(filename):
    path, headers = urlretrieve(url, filename, show_progress_bar)
```

### Unzip

```python
import gzip
import shutil
import os

if os.path.exists(filename):
    print("Unzipping...")
    with gzip.open(filename, 'rb') as f_in:
        with open(filename.rstrip('.gz'), 'wb') as f_out:
            shutil.copyfileobj(f_in, f_out)

    os.remove(filename) # remove gzip file once it is successfully unzipped

    print("Done.")
```
